1.  Explain the difference between Absolute, Relative, and Mixed Cell
Referencing in Excel with examples.

   -> **Relative Cell Reference (A1)**
- Description: The most common type; it adjusts automatically based on the formula's new position.
- Example: If =A1+B1 in cell C1 is copied to C2, it becomes =A2+B2.
- Use Case: Summing different rows of data or multiplying corresponding cells in a range.

**Absolute Cell Reference ($A$1)**
- Description: The cell reference remains constant, no matter where the formula is copied.
- Example: If =B1*$A$1 (where $A$1 holds a discount rate) is copied down, it always refers to $A$1.
- Use Case: Referencing a fixed value like a tax rate, commission percentage, or a specific lookup cell.
- Shortcut: Press the F4 key after selecting the cell reference in the formula bar to toggle to absolute.

**Mixed Cell Reference ($A1 or A$1)**
- Description: Locks either the column or the row, but not both, allowing one part to change and the other to stay fixed.
- $A1 (Absolute Column, Relative Row): Column A stays fixed, but the row number changes when copied.
- A$1 (Relative Column, Absolute Row): Row 1 stays fixed, but the column letter changes when copied.
- Use Case: Creating multiplication tables or calculating areas where one dimension (e.g., width) changes by row and another (e.g., depth) by column.

**Example Scenario: Calculating Area**
- Data: Width in B3, Depth in C2.
- Formula in D3 (Area):
     - =B3*C2 (Copies to D4 as =B4*C3 - Relative)
     - =$B$3*$C$2 (Stays the same - Absolute)
     - =$B3*C$2 (Copies down as =$B4*C$2, across as =$B3*D$2 - Mixed)

3. What are Text Functions in Excel? Mention any five with examples.

   -> Text functions in Excel are formulas designed to manage and manipulate textual data within cells, enabling tasks like joining text, changing case, or extracting parts of a string [1].

Here are five common text functions with examples:

### CONCAT
- Description: Joins two or more text items into one string. It's a modern replacement for the CONCATENATE function [1].
- Example: =CONCAT("First", "Last") returns "FirstLast" [1].
### LEFT
- Description: Extracts a specified number of characters from the beginning (left side) of a text string [1].
- Example: =LEFT("Apple", 3) returns "App" [1].
### RIGHT
- Description: Extracts a specified number of characters from the end (right side) of a text string [1].
- Example: =RIGHT("Apple", 2) returns "le" [1].
### UPPER
- Description: Converts all characters in a text string to uppercase [1].
- Example: =UPPER("hello") returns "HELLO" [1].
### LEN
- Description: Returns the number of characters in a text string, including spaces [1].
- Example: =LEN("Hello World") returns 11 [1]

5. Define the purpose of VLOOKUP and HLOOKUP. How are they different
from XLOOKUP? Which among XLOOKUP and INDEX-MATCH is best while usage?

   -> VLOOKUP searches vertically (column-wise) from the first column, HLOOKUP horizontally (row-wise) from the first row, both returning related data, but XLOOKUP is a modern, superior replacement offering vertical/horizontal search, default exact match, left-to-right search, error handling, and returning multiple values, making XLOOKUP generally better than INDEX-MATCH for simplicity and power unless you need legacy compatibility.

**VLOOKUP & HLOOKUP**
- Purpose: Find a value in a table and return a corresponding value in the same row (VLOOKUP) or column (HLOOKUP).
- VLOOKUP: Searches down the first column, returns from the same row.
- HLOOKUP: Searches across the first row, returns from the same column.
- Limitation: Must search from the first column/row, can't easily search right-to-left or bottom-to-top.
**XLOOKUP vs. VLOOKUP/HLOOKUP**
- Direction: XLOOKUP handles vertical & horizontal, left-to-right & right-to-left.
- Flexibility: Returns entire rows/columns, handles errors, no column index numbers needed.
- Default: Exact match is default (VLOOKUP needs specification).
**XLOOKUP vs. INDEX-MATCH**
- XLOOKUP: Simpler syntax, easier to write, faster for most, built-in error handling, returns multiple results.
- INDEX-MATCH: More complex but powerful for older Excel versions, highly flexible (searches any direction).
- Which is Best?: Use XLOOKUP if available (newer Excel) for ease and features; use INDEX-MATCH or VLOOKUP for compatibility with older Excel.
